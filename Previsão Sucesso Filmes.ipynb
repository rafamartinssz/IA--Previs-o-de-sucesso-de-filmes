# Previsão de sucesso de filmes
Este é um código que realiza a previsão de sucesso de filmes com base no aprendizado de máquina, usando uma base de dados com mais de 800 filmes para o treino dos modelos.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
tabela = pd.read_csv("movie_success_rate.csv")

display(tabela)
```

```python
print(tabela.columns)

print(tabela.info())
```

```python
# Removendo valores NA
print(tabela.isna().sum())
tabela = tabela.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder

# Transformar as colunas de texto em números para o aprendizado da IA
codificador = LabelEncoder()
tabela["Actors"] = codificador.fit_transform(tabela["Actors"])
tabela["Director"] = codificador.fit_transform(tabela["Director"])
```

```python
# Treinando a IA
y = tabela['Success']

x = tabela.drop(['Success', 'Description', 'Genre', 'Title'], axis = 1)

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.25)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

modelos = {
  'Regressão Logistica' : LogisticRegression(),
  'k-NN' : KNeighborsClassifier(),
  'SBM' : SVC(),
  'Árvore de Decisão' : RandomForestClassifier()
}

for nome, modelo in modelos.items():
    modelo.fit(x_treino, y_treino)
    y_pred = modelo.predict(x_teste)
    acuracia = accuracy_score(y_teste, y_pred)
    print(f'{nome} : Acurácia = {acuracia:.4f}')
```
